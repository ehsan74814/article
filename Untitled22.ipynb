```python
# --- Ø³Ù„ÙˆÙ„ Û²: Ú©Ø¯ Ú©Ø§Ù…Ù„ Ùˆ Ù†Ù‡Ø§ÛŒÛŒ Ø¨Ø±Ù†Ø§Ù…Ù‡ ---

import gradio as gr
from PIL import Image
import pytesseract
from transformers import pipeline

print("â³ Ø¯Ø± Ø­Ø§Ù„ Ø¨Ø§Ø±Ú¯Ø°Ø§Ø±ÛŒ Ù…Ø¯Ù„ Ù‡ÙˆØ´ Ù…ØµÙ†ÙˆØ¹ÛŒ...")
ner_pipeline = pipeline(
    "ner",
    model="d4data/biomedical-ner-all",
    aggregation_strategy="simple"
)
print("âœ… Ù…Ø¯Ù„ Ø¨Ø§ Ù…ÙˆÙÙ‚ÛŒØª Ø¨Ø§Ø±Ú¯Ø°Ø§Ø±ÛŒ Ø´Ø¯.")


# --- ØªØ§Ø¨Ø¹ Ø§ØµÙ„ÛŒ Ù¾Ø±Ø¯Ø§Ø²Ø´ (Ø§ØµÙ„Ø§Ø­ Ø´Ø¯Ù‡) ---
def analyze_english_prescription(prescription_image):
    if prescription_image is None:
        return "Please upload an image first.", {}

    try:
        # 1. Ø®ÙˆØ§Ù†Ø¯Ù† Ù…ØªÙ† Ø§Ø² ØªØµÙˆÛŒØ± (OCR)
        extracted_text = pytesseract.image_to_string(prescription_image, lang='eng')

        if not extracted_text.strip():
            return "No text could be read.", {}

        # 2. Ø§Ø³ØªØ®Ø±Ø§Ø¬ Ø§Ø·Ù„Ø§Ø¹Ø§Øª Ø¨Ø§ Ù…Ø¯Ù„ Ù‡ÙˆØ´ Ù…ØµÙ†ÙˆØ¹ÛŒ (NLP)
        ner_results = ner_pipeline(extracted_text)

        # 3. Ø¯Ø³ØªÙ‡â€ŒØ¨Ù†Ø¯ÛŒ Ù†ØªØ§ÛŒØ¬ (Ø¨Ø§ Ø¯Ø³ØªÙ‡â€ŒÙ‡Ø§ÛŒ Ø¬Ø¯ÛŒØ¯)
        # *** Ø§ÛŒÙ†Ø¬Ø§ Ù…Ù‡Ù…â€ŒØªØ±ÛŒÙ† ØªØºÛŒÛŒØ± Ø§Ø³Øª ***
        categorized_results = {
            "Medications": [],
            "Symptoms & Diseases": [],
            "Anatomical Structures": [],
            "Diagnostic Procedures": [],
            "Lab Values": [],
            "Dates": []
        }

        entity_map = {
            "Medication": "Medications",
            "Disease_disorder": "Symptoms & Diseases",
            "Symptom": "Symptoms & Diseases",
            "Biological_structure": "Anatomical Structures",
            # --- Ø¯Ø³ØªÙ‡â€ŒÙ‡Ø§ÛŒ Ø¬Ø¯ÛŒØ¯ Ø¨Ø± Ø§Ø³Ø§Ø³ Ø®Ø±ÙˆØ¬ÛŒ Ø¯ÛŒØ¨Ø§Ú¯ Ø´Ù…Ø§ ---
            "Diagnostic_procedure": "Diagnostic Procedures",
            "Lab_value": "Lab Values",
            "Date": "Dates"
        }

        for entity in ner_results:
            entity_group = entity['entity_group']
            if entity_group in entity_map:
                category = entity_map[entity_group]
                # Ø¬Ù„ÙˆÚ¯ÛŒØ±ÛŒ Ø§Ø² Ø§ÙØ²ÙˆØ¯Ù† Ù…ÙˆØ§Ø±Ø¯ ØªÚ©Ø±Ø§Ø±ÛŒ
                if entity['word'] not in categorized_results[category]:
                    categorized_results[category].append(entity['word'])

        final_results = {k: v for k, v in categorized_results.items() if v}

        return extracted_text, final_results

    except Exception as e:
        return f"An error occurred: {str(e)}", {}


# --- Ø³Ø§Ø®Øª Ø±Ø§Ø¨Ø· Ú©Ø§Ø±Ø¨Ø±ÛŒ Ù†Ù‡Ø§ÛŒÛŒ ---
iface = gr.Interface(
    fn=analyze_english_prescription,
    inputs=gr.Image(type="pil", label="Upload Prescription Image"),
    outputs=[
        gr.Textbox(label="ğŸ“„ Raw Extracted Text (OCR)"),
        gr.JSON(label="ğŸ©º Key Extracted Information (AI)")
    ],
    title="ğŸ‘¨â€âš•ï¸ Smart Prescription Analyzer (Working Version)",
    description="Upload a clear image of an English medical prescription.",
    allow_flagging="never"
)

print("\nğŸš€ Ø¯Ø± Ø­Ø§Ù„ Ø³Ø§Ø®Øª Ù„ÛŒÙ†Ú© Ø¹Ù…ÙˆÙ…ÛŒ Ø¨Ø±Ø§ÛŒ Ø¨Ø±Ù†Ø§Ù…Ù‡...")
iface.launch(share=True)
```
